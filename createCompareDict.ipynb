```python
# coding: utf-8

# In[15]:

import random

def transfer(n):
    '''
    input is a number
    return a string in a card
    '''
    if n>0 and n<9:
        return str(n+1)
    elif n==0:
        return str('A')
    elif n==9:
        return str('T')
    elif n==10:
        return str('J')
    elif n==11:
        return str('Q')
    else:
        return str('K')

def cards_vec():
    '''
    return a vector of 52 cards in poker
    '''
    origina_num=range(52)
    table=[]
    for cardt in origina_num:
        if cardt%4==0:
            table.append(transfer(cardt//4)+'S')
        elif cardt%4==1:
            table.append(transfer(cardt//4)+'H')        
        elif cardt%4==2:
            table.append(transfer(cardt//4)+'D')
        else:
            table.append(transfer(cardt//4)+'C')   
    return table
```

```python
allCards = cards_vec()
```

```python
def make_comparator(comparator5Card):
    def compare(x, y):
        if comparator5Card(x, y):
            return -1
        elif comparator5Card(y, x):
            return 1
        else:
            return 0
    return compare
```

```python
def comparator5Card(a, b):
    a = list(a)
    b = list(b)
    
    better, kind = deal_compare.compare(a, b)
    return a == better
```

```python
deal_compare.compare(list(('AS', 'AH', 'AD', 'AC', '3C')), list(('AS', '9H', 'AD', '8C', '3C')))
```

```python
allCombSorted = sorted(allComb, cmp=make_comparator(comparator5Card))
```

```python
staticCompareDict = {}

sortedFive = sorted(allCombSorted[0])
joinedFive = "".join(sortedFive)
staticCompareDict[joinedFive] = 0
lastFive = sortedFive

rank = 0
for i in xrange(1, len(allCombSorted[0:10])):
    sortedFive = sorted(allCombSorted[i])
    joinedFive = "".join(sortedFive)
    if comparator5Card(sortedFive, lastFive):
        staticCompareDict[joinedFive] = i
        rank = i
    else:
        staticCompareDict[joinedFive] = rank
    lastFive = sortedFive
    
```
